```r
# --- 2. Helper: read one file, clean colnames, add SITE_ID --------------

read_and_clean_site <- function(path) {
  site_id <- file_path_sans_ext(basename(path))
  
  df <- read_csv(path, show_col_types = FALSE)
  
  # Only rename columns that actually exist in this file
  cols_to_rename <- intersect(names(rename_map), names(df))
  
  # dplyr::rename() expects new_name = old_name
  # Our rename_map is old_name = new_name, so invert it here
  rename_vec <- setNames(cols_to_rename, rename_map[cols_to_rename])
  
  df <- df %>%
    rename(!!!rename_vec) %>%
    mutate(SITE_ID = site_id)
  
  df
}
```

```r
# --- 3. Read all site files into a named list ---------------------------

csv_files <- list.files(
  path = "merged_sites_subset",
  pattern = "\\.csv$",
  full.names = TRUE
)
# Now you have: site_list[["DE-Hai"]] etc.
```

```r
output_dir <- "merged_sites_subset_renamed"
dir.create(output_dir, showWarnings = FALSE)
tic("Reading all sites")
walk(csv_files, function(path) {
  site_id <- file_path_sans_ext(basename(path)) |>
  sub("_merged$", "", .)
  site_df <- read_and_clean_site(path)
  out_path <- file.path(output_dir, paste0(site_id, ".csv"))
  write_csv(site_df, out_path)
})
toc()
```
