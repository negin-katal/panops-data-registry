```r
#### datacleaning:

library(data.table)

df <- data
# make sure it's a data.table
setDT(df)

```

```r
# =========================================================
# 1) REPLACE COMMON MISSING VALUE CODES WITH NA
# =========================================================

missing_codes <- c(-9999, -9999.0, -6999, -6999.0, -7999, -7999.0)

num_cols <- names(df)[sapply(df, is.numeric)]

for (col in num_cols) {
  df[get(col) %in% missing_codes, (col) := NA_real_]
}
```

```r
# =========================================================
# 2) IDENTIFY COLUMN GROUPS
# =========================================================

ta_cols  <- grep("^TA_", names(df), value = TRUE)
vpd_cols <- grep("^VPD_", names(df), value = TRUE)
sw_cols  <- grep("^SW_", names(df), value = TRUE)
p_cols   <- grep("^P_", names(df), value = TRUE)

# deadwood / forest / loss columns if present
deadwood_cols <- grep("deadwood", names(df), value = TRUE, ignore.case = TRUE)
forest_cols   <- grep("forest", names(df), value = TRUE, ignore.case = TRUE)
loss_cols     <- grep("loss", names(df), value = TRUE, ignore.case = TRUE)
```

```r
# =========================================================
# 3) RANGE-BASED CLEANING
# =========================================================

# ---- Temperature (deg C) ----
# monthly means / quantiles outside these bounds are treated as impossible
for (col in ta_cols) {
  df[get(col) < -80 | get(col) > 60, (col) := NA_real_]
}

# ---- VPD ----
# should not be negative; upper bound can be adjusted
for (col in vpd_cols) {
  df[get(col) < 0 | get(col) > 20, (col) := NA_real_]
}

# ---- Shortwave radiation ----
# should not be negative
for (col in sw_cols) {
  df[get(col) < 0 | get(col) > 1500, (col) := NA_real_]
}

# ---- Precipitation ----
# should not be negative
for (col in p_cols) {
  df[get(col) < 0, (col) := NA_real_]
}

# ---- Fraction / percent variables ----
# adapt depending on units
# If values are proportions: 0-1
# If values are percentages: 0-100
#
# From your names, these might already be percentages, so check first.
fraction_like_cols <- unique(c(deadwood_cols, forest_cols, loss_cols))

for (col in fraction_like_cols) {
  # first only remove impossible negatives
  df[get(col) < 0, (col) := NA_real_]
}
```

```r
# =========================================================
# 4) OPTIONAL: CHECK FOR REMAINING EXTREME VALUES
# =========================================================

check_extremes <- function(cols, data) {
  if (length(cols) == 0) return(NULL)
  data.table(
    variable = cols,
    min = sapply(cols, function(x) suppressWarnings(min(data[[x]], na.rm = TRUE))),
    max = sapply(cols, function(x) suppressWarnings(max(data[[x]], na.rm = TRUE))),
    n_na = sapply(cols, function(x) sum(is.na(data[[x]])))
  )
}

ta_check  <- check_extremes(ta_cols, df)
vpd_check <- check_extremes(vpd_cols, df)
sw_check  <- check_extremes(sw_cols, df)
p_check   <- check_extremes(p_cols, df)

cat("TA extremes:\n")
print(head(ta_check[order(min)], 20))

cat("\nVPD extremes:\n")
print(head(vpd_check[order(min)], 20))

cat("\nSW extremes:\n")
print(head(sw_check[order(min)], 20))

cat("\nP extremes:\n")
print(head(p_check[order(min)], 20))
```
